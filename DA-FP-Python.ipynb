# **Add URL from github, Merge 12 Dataframe to 1**

```python
## Get and processing data from 12 file.csv on Github ##
path = 'https://raw.githubusercontent.com/KadianAsian/DAFinalProjectData/main/'

## Create a loop to add 12 'tail' of URL into list ##
file = []
for i in range (1,13):
  file.append(f'sales2019_{i}.csv')

## Merge dataframe ##
frames = []

for i in file:
  filepath = path + i
  df1 = pd.read_csv(filepath)
  frames.append(df1)
  df = pd.concat(frames)

df
```

```python
## Dataframe Info ##
df.info()
```

```python
## Descriptive Statistics ##
df.describe()
```

# **Processing Data**

```python
## Add 'Month' Column ##
df['Month'] = df['Order Date'].str[0:2]

## Check ['Month'] column ##
print(set(df['Month']))
```

```python
## Checking NaN values ##
check_NaN = df[df['Month'].isna()]
check_NaN.head()
```

```python
## Checking 'Or' values ##
check_Or = df[df['Month'] == 'Or']
check_Or.head()
```

```python
## Find % of missing data ##
total_row = len(df.index)
total_nan = df['Month'].isna().sum()
percent_nan = (total_nan / total_row) * 100
percent_nan
# Missing data ~ 0.3% ===> delete
```

```python
## Clear NaN values ##
df = df.dropna(how='all')
df.isna().sum()
```

```python
## Clear 'Or' values ##
df = df[df['Month'] != 'Or']
df[df['Month'] == 'Or'].count()
```

# **Convert Dtype**

```python
df.info()
```

```python
## Convert dtype's ['Quantity Ordered'] column to int
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], downcast = 'integer')
df['Price Each'] = pd.to_numeric(df['Price Each'], downcast = 'integer')

## Convert dtype's ['Order Date'] columns to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

df.info()
```

## **Add columns ['City'], ['Sales'], ['Hour'] & moving position**

```python
# Add ['City'] column
address_to_city = lambda address:address.split(', ')[1]
df['City'] = df['Purchase Address'].apply(address_to_city)

## Add ['Hours'] columns
df['Hours'] = df['Order Date'].dt.hour

## Add ['Sales'] column ##
df['Sales'] = df['Quantity Ordered'] * df['Price Each']

df.head()
```

```python
## Moving ['Sales'] & ['Month'] & ['Hours'] columns position ##

# Moving ['Sales'] #
moving_column = df.pop('Sales')
insert_column = df.insert(4, 'Sales', moving_column)

# Moving ['Hours'] #
moving_hours = df.pop('Hours')
insert_hours = df.insert(7, 'Hours', moving_hours)

# Moving ['Month'] #
moving_month = df.pop('Month')
insert_month = df.insert(6, 'Month', moving_month)

df.head()
```

# **What city has the best sales?**

```python
## Sales value of each city ##
sv_city = df.groupby('City')['Sales'].sum()

## Print the conclude ##
print(f'The city has the best sales is {sv_city.idxmax()}, with {sv_city.max()} USD')

sv_city
```

```python
## Plot chart ##
cities = sv_city.index
salescities_plot = plt.bar(x=cities, height=sv_city)
plt.title('Sales Of Cities', fontweight='bold')
plt.xlabel('Cities', fontweight='bold')
plt.ylabel('Sales in USD', fontweight='bold')
plt.xticks(x=cities, rotation=45, size=8)
salescities_plot[7].set_color('green')
plt.show()
```

# **The price of the best-selling product?**

```python
## Total quantity ordered of all products ##
product_total_order = df.groupby('Product')['Quantity Ordered'].sum()

## Average of each product ##
prices = df.groupby('Product')['Price Each'].mean()

## Get name of each product ##
product = prices.index

## Print the conclude ##
for i in prices.index:
  if i == product_total_order.idxmax():
    name_product = i

avg_pp = df[df['Product'] == name_product]['Price Each'].mean()

print(f'The best-selling product is {product_total_order.idxmax()}, have average price is {round(avg_pp,2)}')
```

```python
## Plot ##
x = product
y1 = product_total_order
y2 = prices
fig, ax1 = plt.subplots()
fig.suptitle('Quantity Ordered & Price Of All Products',  fontweight='bold', fontsize=15)

ax2 = ax1.twinx()
ax1.bar(x, y1)
ax2.plot(x, y2, 'm-')

ax1.set_xticklabels(product, rotation=90, size=8)
ax1.set_xlabel('Products', fontweight='bold')
ax1.set_ylabel('Total Quantity Ordered', color='b', fontweight='bold')
ax2.set_ylabel('AVG price (in USD)', color='m', fontweight='bold')

plt.show()
```

# **What was the best month for sale?**


```python
## Monthly sales ##
sales_value = df.groupby(by='Month')['Sales'].sum()

## Get month ##
months = sales_value.index

## Print conclude ##
print(f'The best month for sale is {sales_value.idxmax()}')
```

```python
## Plot ##
salesmonth_plot = plt.bar (x=months, height=sales_value)
plt.title('Monthly Sales',  fontweight='bold')
plt.xticks(months)
plt.xlabel('Months', fontweight='bold')
plt.ylabel('Sales in USD', fontweight='bold')
salesmonth_plot[11].set_color('green')
plt.show()
```

# **What time was most orders in day!**

```python
## Get quantity ordered each hour in day ##
sv_hour = df.groupby('Hours')['Sales'].count()

## Print conclude ##
print(f"The time was most orders in day is {sv_hour.idxmax()} o'clock!")
```

```python
## Plot ##
hour = range(0,24)
sv_hour_plot = plt.plot(hour, sv_hour)
plt.title('Number Of Orders At Times Of The Day', fontweight='bold')
plt.xticks(hour, size=8)
plt.xlabel('Hours', fontweight='bold')
plt.ylabel('Order in hour', fontweight='bold')
plt.grid(which='major')
plt.show()
```

# **What products are most often sold together?**

```python
## Find orders placed at the same time ##
df_dup = df[df['Order ID'].duplicated(keep=False)]

## Create new dataframe with products was ordered together ##
groupProduct = lambda product: ', '.join(product)
df_dup['All Products'] = df_dup.groupby('Order ID')['Product'].transform(groupProduct)
df_dp = df_dup[['Order ID','All Products']].drop_duplicates()

## Get top 10 products are most often sold together ##
top10_dp = df_dp['All Products'].value_counts().head(10)
top10_dp

## Print conclude ##
print(f'2 products are most often sold together is {top10_dp.idxmax()}')
```

```python
## Plot Chart ##
products = top10_dp.index
top10_dp_plot = plt.barh(y=products, width=top10_dp)
plt.title('Top 10 Products Are Most Often Sold Together', fontweight='bold')
plt.xlabel('Order Placed Together', fontweight='semibold')
plt.ylabel('Products', fontweight='semibold')
plt.xticks(x=products, size=8)
top10_dp_plot[1].set_color('g')
top10_dp_plot[0].set_color('g')
plt.show()
```
